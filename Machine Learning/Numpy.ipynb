# Numpy Array Operations  

## 1. Basic Operations  
- **a.** \( A - b \)  
- **b.** \( b^2 \)  
- **c.** \( A^{c} \)  
- **d.** \( A \cdot B \) (Element-wise multiplication)  
- **e.** \( A.dot(B) \)  
- **f.** \( A = B \)  

## 2. Unary Operations  
- **a.** A.min() and A.max()  
- **b.** A.sum(axis=0) 

## 3. Universal Functions  
- **a.** np.exp(A)  
- **b.** np.sqrt(A)
- **c** np.sin(A)
