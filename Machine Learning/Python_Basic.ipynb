## Loops in Python
**Need**

- Repeat code block.
- Iteration: Execute same code repeatedly.
 
**While Loop**
- Executes while condition true.

**For Loop**
- Executes for each item in iterable.

````python
#for Loop and While Loop
for i in range(1,4):
    print(i)  
i=1
while i<=3:
    print(i)  
    i+=1
````

## Loop Control Statements
**Break:** Ends loop immediately.

**Continue:** Skips to next iteration.

**Pass:** Placeholder; does nothing.

````python
#break Statement
for i in range(1,6):
    if i==3:
        break
    print(i)    
#continue Statement
for i in range(1,6):    
    if i==3:
        continue
    print(i)
#pass Statement
for i in range(1,6):    
    if i==3:
        pass
    print(i)
````

# Lists vs Tuples  

## Syntax:  
- **Lists:** `[]`  
- **Tuples:** `()`  

## Mutability:  
- **Lists:** Mutable  
- **Tuples:** Immutable  

## Speed:  
- **Lists:** Slower (mutable)  
- **Tuples:** Faster (immutable)  

## Memory:  
- **Lists:** Higher  
- **Tuples:** Lower  

## Functionality:  
- **Both:** Indexing, slicing  
- **Lists:** More methods  

## Error-Prone:  
- **Lists:** Modifiable  
- **Tuples:** Safer  

## Use Case:  
- **Lists:** Dynamic  
- **Tuples:** Static

````python
import time 
L = list(range(100000000))
T = tuple(range(100000000))

# List timing
start = time.time()
for i in L:
  i*5
print('List time', time.time()-start)

# Tuple timing
start = time.time()
for i in T:
  i*5
print('Tuple time', time.time()-start)
````

### Q. Why use Tuples?
- Immutable; prevents changes.
- Ensures data integrity.
- Use for fixed collections.
- Example:

  ```python
  college_database = ('CS', 'Math', 'Physics')
  # college_database[0] = 'Electronics'  # TypeError
  ```
- Use for static data; lists for mutable.
  

````python
# tuple unpacking
a, b, c = (1, 2, 3)
print(a, b, c)
````

````python
a, b = (1, 2, 3)
print(a, b)
````

````python
a = 1
b = 2
a, b = b, a
print(a, b)
````
